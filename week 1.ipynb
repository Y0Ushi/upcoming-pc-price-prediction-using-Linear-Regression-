# Linear Regression

```python
from sklearn import linear_model
```

# we already know :

       y = M1X1 + M2X2 + M3X3 +......+MnXn + c
       
       X1 = Item ID
       x2 = Ram_Gb
       x3 = StoraGe_Gb	
       X4  = processorSpeed
       y = Price_Rs
       
       Price_Rs = M1*Item ID + M2*Ram_Gb + M3* StoraGe_Gb + M4*processorSpeed + intercept

```python
 reg = linear_model.LinearRegression()
```

```python
 reg
```

```python
 reg.fit(ds[['Item ID' , 'Ram_Gb' , 'StoraGe_Gb' , 'processorSpeed']],ds.Price_Rs)
```

```python
reg.predict([[32 , 32 , 2048 , 4.50 ]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
